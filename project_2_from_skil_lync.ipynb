```python
import pandas as pd 
df = pd.read_csv("C:\\Users\\vt645\\Downloads\\customer_details.csv")
df.columns=['customer_id','gender','age','driving_licence_present','region_code','perviously_insured','vehicle_age','vehicle_damage']
df
```

```python
#Generate a summary of count of all the null values column wise
df.isnull().sum()
```

```python
#Drop Null values for customer_id because central tendencies for id’s is not feasible.
df.dropna(subset=['customer_id'],inplace=True)
df
```

```python
#Replace all null values for numeric columns by mean
mean1= df.customer_id.mean()
mean2= df.driving_licence_present.mean()
mean3= df.region_code.mean()
mean4= df.perviously_insured.mean()
mean5 = df.age.mean()
```

```python
df['customer_id'].fillna(mean1,inplace=True)
df['driving_licence_present'].fillna(mean2,inplace=True)
df['region_code'].fillna(mean3,inplace=True)
df['perviously_insured'].fillna(mean4,inplace=True)
df['age'].fillna(mean5,inplace=True)
```

```python
# Replace all null values for Categorical value by mode.
mode_value1 = df['gender'].mode()[0]
mode_value2 = df['vehicle_damage'].mode()[0]
mode_value3 = df['vehicle_age'].mode()[0]


print(mode_value1)
print(mode_value2)
print(mode_value3) 

df['gender'].fillna(mode_value1,inplace=True)
df['vehicle_damage'].fillna(mode_value1,inplace=True)
df['vehicle_age'].fillna(mode_value1,inplace=True)

```

```python
'''Sure! The formula for calculating the first quartile (Q1) is as follows:

Arrange the data set in ascending order.
Calculate the position of the first quartile (Q1) using the formula (n + 1) / 4, where n is the number of data points in the set.
If (n + 1) / 4 is not an integer, round up to the nearest integer and use that as the position of Q1.
Find the value at the position of Q1 in the data set.
In Hindi, you can say “पहला चौथाई भाग” to refer to Q1'''
```

```python
#Generate a summary of count of all the outliers column wise

q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
Iqr = q3-q1
lower_bound = q1-(1.5 * Iqr)
upper_bound =  q3+(1.5 * Iqr)
#print(f"Lower bound: {lower_bound}")
#print(f"Upper bound: {upper_bound}")

outlier = ((df<lower_bound) | (df>upper_bound))
print(outlier.sum())
```
